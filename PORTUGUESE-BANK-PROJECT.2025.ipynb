```python
#Count plot
# Create a new custom color palette with blues and greens
custom_palette = ["#3498DB", "#2ECC71", "#1ABC9C"]  # Blue, Green, Teal

# Set up the figure with appropriate number of rows and columns
fig, axes = plt.subplots(5, 2, figsize=(20, 25), facecolor='white')
axes = axes.flatten()  # Flatten to make iteration easier

# Loop through each categorical column and create count plots
for plotnumber, column in enumerate(cat_data.columns):
    ax = axes[plotnumber]
    sns.countplot(x=cat_data[column], hue=data['y'], ax=ax, palette=custom_palette)
    
    # Annotate each bar with its corresponding count value, handling NaNs safely
    for p in ax.patches:
        height = p.get_height()
        if not np.isnan(height):  # Check if height is not NaN
            ax.annotate(f'{int(height)}', 
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=12, color='black', 
                        xytext=(0, 5), textcoords='offset points')
        else:
            ax.annotate('N/A', 
                        (p.get_x() + p.get_width() / 2., 0),
                        ha='center', va='bottom', fontsize=12, color='red', 
                        xytext=(0, 5), textcoords='offset points')

    # Set axis labels and improve readability
    ax.set_xlabel(column, fontsize=16)
    ax.set_ylabel('Count', fontsize=16)
    ax.set_title(f'Distribution of {column}', fontsize=18)
    ax.tick_params(axis='x', rotation=45)

# Remove empty subplots if cat_data has fewer than expected columns
if len(cat_data.columns) < len(axes):
    for i in range(len(cat_data.columns), len(axes)):
        fig.delaxes(axes[i])

# Adjust layout to prevent overlap and optimize spacing
plt.tight_layout()
plt.show()
```

From above count plot analysis highlights clear trends in customer behavior, segmented by jobs, marital status, education, loans, contact methods, and previous outcomes. Insights from these patterns can guide targeted marketing strategies, such as focusing on higher-education customers, leveraging cellular contact methods, and improving engagement during active months like May. Addressing imbalances in certain groups (e.g., "Unknown" statuses) and analyzing patterns across occupations can further refine strategies for improved results.

# *Model Creation*

  ### ***Logistic Regression***

```python
# import libraries
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_smote,y_smote)
```

```python
y_pred=log.predict(x_test)
```

```python
#Model evaluaion
```

```python
#import Libraries
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,classification_report
```

```python
#Accuracy score
lr=accuracy_score(y_test,y_pred)
print(lr)
```

```python
# classification_report
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

# Plot the ROC curve with new colors and styles
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='#2ECC71', lw=2, label='ROC curve (AUC = {:.2f})'.format(roc_auc))  # Green curve
plt.plot([0, 1], [0, 1], color='#E74C3C', lw=2, linestyle='--', label='Random (Chance)')  # Red dashed line

# Set plot limits and labels
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14, fontweight='bold', color='#34495E')
plt.ylabel('True Positive Rate', fontsize=14, fontweight='bold', color='#34495E')
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16, fontweight='bold', color='#2C3E50')

# Add legend with a customized position
plt.legend(loc='lower right', fontsize=12)

# Add grid for better visualization
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.show()
```

Our logistic regression model demonstrates strong performance, with an accuracy score of 83% and an ROC AUC of 0.83, indicating its effectiveness.

# *Challenges during the project*

### *1. Imbalanced Target Variable*

**Challenge**:
The dataset was highly imbalanced, with the majority of target values being "No" (non-subscribers), making it difficult for models to predict the minority class ("Yes").

**Solution**:
Applied SMOTE (Synthetic Minority Oversampling Technique) to oversample the minority class.
Used class weights in models to give more importance to the minority class.
Evaluated models using precision, recall, F1-score, and ROC AUC instead of relying solely on accuracy.

### *2. Outliers in Numerical Data*

**Challenge**:
Numerical columns like 'duration', 'campaign', and 'pdays' contained extreme outliers that distorted model performance.

**Solution**:
Detected outliers using box plots and histograms.
Applied capping (e.g., limiting extreme values to a certain percentile) to reduce the impact of outliers.
Used log transformations for skewed variables like 'duration' and 'pdays' to normalize their distributions.

### *3. Skewed Distributions*

**Challenge**:
Variables like 'pdays' and 'previous' were highly skewed, with a majority of zero values and a few extreme values.

**Solution**:
Grouped skewed variables into bins to simplify patterns and improve model interpretability.

### *4. High Cardinality in Categorical Variables*

**Challenge**:
Features like 'job' and 'education' had many unique categories, increasing the complexity of encoding and risk of overfitting.

**Solution**:
Used frequency encoding for high-cardinality variables to reduce feature space.
Applied one-hot encoding for low-cardinality variables to retain interpretability.

### *5. Missing or Unknown Data*

**Challenge**:
Many categorical variables, such as 'marital', 'education', and 'default', contained "Unknown" values.

**Solution**:
Treated "Unknown" as a separate category to preserve its informational value.
For numerical features, used mean or median imputation to fill missing values.
